## VI. conexión con chromeDB

```python
import chromadb
from sentence_transformers import SentenceTransformer

# 1. Reconectar a la base de datos que ya creaste
cliente_db = chromadb.PersistentClient(path="./vectordb_rioja")
coleccion = cliente_db.get_collection("normativa")
print(f"✅ ChromaDB conectado — {coleccion.count()} chunks disponibles")

# 2. Modelo de embeddings para vectorizar las preguntas
embedding_model = SentenceTransformer("BAAI/bge-m3")
print("✅ Modelo de embeddings listo")

# 3. Función de búsqueda
def buscar_contexto(pregunta, n_resultados=4):
    vector = embedding_model.encode([pregunta]).tolist()
    resultados = coleccion.query(query_embeddings=vector, n_results=n_resultados)

    fuentes = set(m['fuente'] for m in resultados['metadatas'][0])
    print(f"   📎 Fuentes: {fuentes}")

    return "\n\n---\n\n".join(resultados['documents'][0])
```

### VIII.I Funciones auxiliares de la interfaz

```python
if IN_COLAB:
    os.system("wget -q https://raw.githubusercontent.com/Kobakk/Modelo_AI_Agricultura/refs/heads/main/limpieza.py -O limpieza.py")

from limpieza import *

# pdf -> vectorización -> add in ChromaDB
def procesar_y_vectorizar_pdf(archivo_pdf):
    if archivo_pdf is None:
        return "No has subido ningún PDF."

    ruta_pdf = str(archivo_pdf) # Gradio guarda el archivo en una ruta temporal
    nombre = os.path.basename(ruta_pdf).replace('.pdf', '')

    # 1. Extraer texto del PDF (función de limpieza.py)
    try:
        paginas = extraer_texto_crudo(ruta_pdf)
    except Exception as e:
        return f"Error al leer el PDF: {e}"

    # 2. Comprobar si tiene texto
    texto_total = sum(p['num_caracteres'] for p in paginas)
    if texto_total < 500:
        return f"'{nombre}.pdf' parece escaneado (muy poco texto). Necesitaría OCR."

    # 3. Limpiar y crear chunks
    todos_los_chunks = []
    for pagina in paginas:
        texto_limpio = limpiar_pagina(pagina['texto_crudo'])
        chunks = crear_chunks(texto_limpio, nombre, pagina['pagina'])
        todos_los_chunks.extend(chunks)

    if not todos_los_chunks:
        return f"No se generaron chunks de '{nombre}.pdf'."

    # 4. Vectorizar e insertar en ChromaDB
    textos = [c["texto"] for c in todos_los_chunks]
    vectores = embedding_model.encode(textos, show_progress_bar=False).tolist()

    coleccion.add(
        documents=textos,
        embeddings=vectores,
        metadatas=[{"fuente": c["fuente"], "pagina": c["pagina"]} for c in todos_los_chunks],
        ids=[f"{nombre}_chunk_{i}" for i in range(len(textos))]
    )

    return f"'{nombre}.pdf' — {len(textos)} chunks añadidos. Total en BD: {coleccion.count()}"
```

```python
def chat_fn(mensaje, archivo_pdf, historial):
    """
    Función principal del chat.
    1. Si hay un PDF adjunto, lo procesa primero
    2. Busca contexto en ChromaDB
    3. Genera respuesta con Gemma
    """
    estado_pdf = ""

    # Si el usuario ha adjuntado un PDF, procesarlo antes de responder
    if archivo_pdf is not None:
        estado_pdf = procesar_y_vectorizar_pdf(archivo_pdf)

    if not mensaje.strip():
        return historial, estado_pdf or "Escribe una pregunta."

    # 1. Buscar los 4 chunks más relevantes
    vector = embedding_model.encode([mensaje]).tolist()
    resultados = coleccion.query(query_embeddings=vector, n_results=4)
    contexto = "\n\n---\n\n".join(resultados["documents"][0])
    fuentes = list(set(m["fuente"] for m in resultados["metadatas"][0]))

    # 2. Construir prompt RAG
    prompt = f"""Eres un asistente experto legal y administrativo, especializado en agricultura.
Usa ÚNICAMENTE el siguiente contexto extraído del boletín oficial para responder de forma clara.
Si la respuesta no se encuentra en el contexto, responde: "No tengo suficiente información en el documento para responder a esto".

Contexto:
{contexto}

Pregunta: {mensaje}"""

    # 3. Generar respuesta con Gemma
    messages = [{"role": "user", "content": prompt}]
    prompt_formateado = tokenizer.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True
    )

    inputs = tokenizer(prompt_formateado, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=500,
        temperature=0.1,
        do_sample=True
    )
    input_length = inputs["input_ids"].shape[-1]
    respuesta = tokenizer.decode(outputs[0][input_length:], skip_special_tokens=True)

    # 4. Montar texto de fuentes
    fuentes_texto = "Fuentes: " + " · ".join(fuentes) if fuentes else ""
    if estado_pdf:
        fuentes_texto = estado_pdf + "\n\n" + fuentes_texto

        historial = historial + [(mensaje, respuesta)]
    return historial, fuentes_texto 
```

```python
def clear_chat():
    return [], ""
```

### VIII.II - Diseño de interfaz

```python
#IN_COLAB = "google.colab" in sys.modules

with gr.Blocks(
    title="AgroConsulta RAG",
    theme=gr.themes.Soft(
        primary_hue=gr.themes.colors.green,
        secondary_hue=gr.themes.colors.orange,
        font=[gr.themes.GoogleFont("DM Sans"), "sans-serif"],
    ),
    css="""
        .source-box {
            background: #f0f7e6;
            border-left: 4px solid #7a9e3b;
            padding: 10px 14px;
            border-radius: 6px;
            font-size: 0.85em;
            color: #4a6741;
        }
        footer { display: none !important; }
        #col-central { max-width: 820px; margin: 0 auto; }
        #row_1 { align-items: center; }
    """,
) as demo:

    with gr.Column(elem_id="col-central"):

        gr.Markdown("## 🍇 AgroConsulta RAG\nConsulta boletines oficiales de agricultura de La Rioja")

        chatbot = gr.Chatbot(
            label="",
            height=460,
            bubble_full_width=False,
            #show_copy_button=True,
        )

        with gr.Row(elem_id="row_1"):
            pdf_upload = gr.UploadButton(
                "+",
                file_types=[".pdf"],
                file_count="single",
                scale=0,
                min_width=50,
                variant="secondary",
            )
            msg_input = gr.Textbox(
                placeholder="Escribe tu pregunta… (Enter para enviar)",
                label="",
                scale=5,
                lines=2,
                max_lines=5,
                autofocus=True,
            )
            send_btn = gr.Button("Enviar", variant="primary", scale=1)

        sources_box = gr.Markdown(elem_classes=["source-box"])
        clear_btn = gr.Button("Nueva conversación", variant="stop", size="sm")


    # EVENTOS DE LA INTERFAZ

    # Variable de estado para guardar el PDF subido
    pdf_state = gr.State(value=None)

    # Cuando el usuario pulsa "+" y sube un PDF, se guarda en el estado
    pdf_upload.upload(
        fn=lambda f: f,
        inputs=[pdf_upload],
        outputs=[pdf_state],
    )

    # Enviar con botón
    send_btn.click(
        fn=chat_fn,
        inputs=[msg_input, pdf_state, chatbot],
        outputs=[chatbot, sources_box],
    ).then(lambda: (None, ""), outputs=[pdf_state, msg_input])

    msg_input.submit(
        fn=chat_fn,
        inputs=[msg_input, pdf_state, chatbot],
        outputs=[chatbot, sources_box],
    ).then(lambda: (None, ""), outputs=[pdf_state, msg_input])

    clear_btn.click(fn=clear_chat, outputs=[chatbot, sources_box])
```
